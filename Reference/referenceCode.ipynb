pandas: 데이터 분석을 위한 라이브러리, 주로 데이터프레임 생성 및 조작에 사용  
numpy: 수치 계산 및 배열/행렬 처리를 위한 라이브러리  
matplotlib.pyplot: 데이터 시각화를 위한 라이브러리, 그래프(라인, 바, 히스토그램 등) 작성에 사용  
train_test_split: 데이터를 학습용과 테스트용으로 나누는 함수  
StandardScaler: 데이터 표준화(평균 0, 표준편차 1로 변환) / LabelEncoder: 범주형 데이터를 숫자로 변환  
KNeighborsClassifier: K-최근접 이웃(KNN) 분류 알고리즘을 제공하는 머신러닝 모델  
seaborn: 고급 통계 시각화 도구, matplotlib 기반으로 다양한 통계적 그래프를 쉽게 생성  


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import seaborn as sns
````

````python

df = pd.read_csv("C:/AI_File/car_evaluation.csv", header = None)
df
````

````python
df.columns=['price', 'maint', 'doors', 'persons', 'lug_capacity', 'safety',
       'output']
df
````

````python
df.columns
````

````python
# 결측치 확인
print("\n 결측값 확인:")
print(df.isnull().sum())
````

````python
columns = ['price', 'maint', 'doors', 'persons', 'lug_capacity', 'safety','output']

label_encoders = {}
for column in columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])


df
````

````python
df['output'].value_counts()
````

````python
df.drop(columns=[], inplace=True)
# 컬럼 제거
df
````

````python
X = df.drop('output', axis=1).values
y = df['output'].values
````

```
X = df.drop('output', axis=1).values
```
1. df에서 'output' 컬럼을 제외한 나머지 컬럼들만 가져와서 X에 저장.  
2. 즉, 입력 데이터(특징 값) 만 추출하는 것.
3. .values를 붙여서 데이터프레임(df) 을 넘파이 배열로 변환.    

 
- axis(방향을 뜻하는 옵션)

  -axis=0 : 행(row) 기준  
  -axis=1 : 열(column) 기준

---
```
y = df['output'].values
```
1. df에서 'output' 컬럼만 선택해서 y에 저장. 
2. 목표 값(라벨, 정답 데이터) 만 추출하는 것.

````python
X
````

````python
y
````

````python
# 데이터 분할 (80% 훈련, 20% 테스트)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
````

X_train: 학습용 특징 데이터  
X_test: 테스트용 특징 데이터  
y_train: 학습용 라벨(정답)  
y_test: 테스트용 라벨(정답)

````python
# 표준화 (Scaling)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
````

````python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)

print("\nDecision Tree Model")
print(f"Accuracy: {accuracy_score(y_test, dt_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, dt_pred))
print("- Confusion Matrix -")
print (confusion_matrix(y_test, dt_pred))
````

````python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

print("\nRandom Forest Model")
print(f"Accuracy: {accuracy_score(y_test, rf_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, rf_pred))
print("- Confusion Matrix -")
print (confusion_matrix(y_test, rf_pred))
````

````python
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)

print("\nSupport Vector Machine Model")
print(f"Accuracy: {accuracy_score(y_test, svm_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, svm_pred))
print("- Confusion Matrix -")
print (confusion_matrix(y_test, svm_pred))
````

````python
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)

print("\nLogistic Regression Model")
print(f"Accuracy: {accuracy_score(y_test, lr_pred):.4f}")
print("Classification Report:")
print(classification_report(y_test, lr_pred))
print("- Confusion Matrix -")
print (confusion_matrix(y_test, lr_pred))
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
````

````python
model = LinearRegression()
model.fit(X_train, y_train)
ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

````

````python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
````

````python
model = RandomForestRegressor()
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
````

````python
model = SVR(kernel='linear')
model.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y_test))
````
